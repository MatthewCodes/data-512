### This cell imports necessary packages, creates a function to call APIs 

```python
import json
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def api_call(endpoint,parameters):
    call = requests.get(endpoint.format(**parameters), headers=headers)
    response = call.json()
    
    return response


headers = {
    'User-Agent': 'https://github.com/MatthewCodes',
    'From': 'mrr313@uw.edu'
}
```

### Endpoints for the API calls

```python
endpoint_legacy = 'https://wikimedia.org/api/rest_v1/metrics/legacy/pagecounts/aggregate/{project}/{access-site}/{granularity}/{start}/{end}'

endpoint_pageviews = 'https://wikimedia.org/api/rest_v1/metrics/pageviews/aggregate/{project}/{access}/{agent}/{granularity}/{start}/{end}'
```

### Creates JSON files for all the access sites for the Pagecount API (desktop, mobile)

```python
param_legacy = ['desktop', 'mobile']

for site in param_legacy:
    # SAMPLE parameters for getting aggregated legacy view data 
    # see: https://wikimedia.org/api/rest_v1/#!/Legacy_data/get_metrics_legacy_pagecounts_aggregate_project_access_site_granularity_start_end
    example_params_legacy = {"project" : "en.wikipedia.org",
                     "access-site" : site+'-site',
                     "granularity" : "monthly",
                     "start" : "2001010100",
                    # for end use 1st day of month following final month of data
                     "end" : "2021010100"
                        }

    example_monthly_legacy = api_call(endpoint_legacy, example_params_legacy)
    dates = []
    
    # This gets all the dates and sorts them from oldest to most recent
    for item in example_monthly_legacy['items']:
        date = item['timestamp'][:6]
        dates.append(date)
    date_set = sorted(set(dates))
    
    # Creates CSV file
    with open('pagecounts_'+site+'_'+date_set[0]+'-'+date_set[-1]+'.json', 'w') as outfile:
        json.dump(example_monthly_legacy, outfile)
```

### Creates JSON files for all the access sites for the Pageview API (desktop, mobile-web, mobile-app)

```python
param_queries = ['desktop', 'mobile-web', 'mobile-app']

for site in param_queries:
    # SAMPLE parameters for getting aggregated current standard pageview data
    # see: https://wikimedia.org/api/rest_v1/#!/Pageviews_data/get_metrics_pageviews_aggregate_project_access_agent_granularity_start_end
    example_params_pageviews = {"project" : "en.wikipedia.org",
                        "access" : site,
                        "agent" : "user",
                        "granularity" : "monthly",
                        "start" : "2001010100",
                        # for end use 1st day of month following final month of data
                        "end" : '2021010100'
                            }

    example_monthly_pageviews = api_call(endpoint_pageviews, example_params_pageviews)
    dates = []
    
    # This gets all the dates and sorts them from oldest to most recent
    for item in example_monthly_pageviews['items']:
        date = item['timestamp'][:6]
        dates.append(date)
    date_set = sorted(set(dates))
    
    # Creates CSV file
    with open('pageviews_'+site+'_'+date_set[0]+'-'+date_set[-1]+'.json', 'w') as outfile:
        json.dump(example_monthly_pageviews, outfile)
```

### Gets all the entries from the pageviews mobile-app file and puts them in a dictionary

```python
json_dic = dict()

entries = []
with open('pageviews_mobile-app_201507-202009.json') as json_file:
    data = json.load(json_file)
    
    for entry in data['items']:
        item = dict()
        
        item['project'] = entry['project']
        item['access'] = 'mobile'
        item['agent'] = entry['agent']
        item['granularity'] = entry['granularity']
        item['timestamp'] = entry['timestamp']
        item['views'] = entry['views']
        entries.append(item)
        
json_dic['items'] = entries
```

### Combines the entries from the mobile-web json file with the mobile-app entries in order to make a single mobile json file for pageviews (pageviews_mobile_201507-202009.json)

```python

with open('pageviews_mobile-web_201507-202009.json') as json_file:
    data = json.load(json_file)
    
    for entry in data['items']:
        time = entry['timestamp']
        
        # If we found that this date has already been added, just add the views to the existing entry
        found = False
        for item in json_dic['items']:
            if item['timestamp'] == time:
                found = True
                item['views'] += entry['views']
                
        # If this date has not been added already create a new entry for it
        if found == False:
            new_entry = dict()
            new_entry['project'] = entry['project']
            new_entry['access'] = 'mobile'
            new_entry['agent'] = entry['agent']
            new_entry['granularity'] = entry['granularity']
            new_entry['timestamp'] = entry['timestamp']
            new_entry['views'] = entry['views']
            
            json_dic['items'].append(new_entry)
            
# Creates CSV file
with open('pageviews_mobile_201507-202009.json', 'w') as outfile:
    json.dump(json_dic, outfile)
```

### Combines all the files into one dictionary in the format of apiName_platform_year_month: amount of views

```python
data_dic = dict()
files = [
'pagecounts_desktop_200712-201608.json',
'pagecounts_mobile_201410-201608.json',
'pageviews_desktop_201507-202009.json',
'pageviews_mobile_201507-202009.json']

# For every entry in every file create an item in the data_dic dictionary
for file in files:
    with open(file) as json_file:
        data = json.load(json_file)
        actual_data = data['items']
        api = file.split("_")[0][:-1]
        platform = file.split("_")[1]
        for entry in actual_data:
            
            year = entry['timestamp'][:4]
            month = entry['timestamp'][4:6]
            try:
                data_dic[api+"_"+platform+"_"+year+"_"+month] = entry['count']
            except:
                data_dic[api+"_"+platform+"_"+year+"_"+month] = entry['views']
```

### Turns dictionary into a matrix with columns: ['year', 'month', 'pagecount_all_views', 'pagecount_desktop_views', 'pagecount_mobile_views', 'pageview_all_views', 'pageview_desktop_views', 'pageview_mobile_views']

```python
years = [str(i) for i in range(2007,2021)]
months= ['0'+str(i) for i in range(1,10)] + [str(i) for i in range(10,13)]

data = []
for year in years:
    for month in months:
        line = [year, month, 0, 0, 0, 0, 0, 0]
        for key, value in data_dic.items():
            info = key.split("_")
            
            # What these next if statements do is fill the above line variable with the corresponding number of views 
            # Given the api, year, month, and platform
            if info[2] == year and info[3] == month:
                if info[0] == 'pagecount':
                    
                    # pagecount desktop
                    if info[1] == 'desktop':
                        line[3] = value
                    # pagecount mobile
                    else:
                        line[4] = value
                else:
                    # pageview desktop
                    if info[1] == 'desktop':
                        line[6] = value
                    # pageview mobile
                    else:
                        line[7] = value
        line[2] = line[3] + line[4]
        line[5] = line[6] + line[7]
        data.append(line)
        
processed_data = data[11:-3]
```

### Turns the data matrix into a dataframe and the en-wikipedia_traffic_200712-202008.csv file

```python

data = {'year': [line[0] for line in processed_data],
        'month': [line[1] for line in processed_data],
        'pagecount_all_views': [line[2] for line in processed_data],
        'pagecount_desktop_views': [line[3] for line in processed_data],
        'pagecount_mobile_views': [line[4] for line in processed_data],
        'pageview_all_views': [line[5] for line in processed_data],
        'pageview_desktop_views': [line[6] for line in processed_data],
        'pageview_mobile_views': [line[7] for line in processed_data]
       }

df_marks = pd.DataFrame(data)
df_marks.to_csv('en-wikipedia_traffic_200712-202008.csv', index=False)
```




### Takes the previously created dataframe drops the year and month then converts it to a numpy array which is then fed into another dataframe with the appropriate frequency and columns (also replaces 0 with nan)

```python
series = df_marks.iloc[:,2:].to_numpy()
df = pd.DataFrame(series, index=pd.date_range(start='12/1/2007', end='10/1/2020', freq='M'), columns=['pagecount_all_views', 'pagecount_desktop_views', 'pagecount_mobile_views', 'pageview_all_views', 'pageview_desktop_views', 'pageview_mobile_views'])
df.replace(0, np.nan, inplace=True)
```

### Plots the dataframe and saves the plot as A1_plot.jpeg

```python
plt.close('all')

plt.figure(figsize=(16,10))
plt.plot(df.index, df.iloc[:,1], 'g--',
         df.index, df.iloc[:,0], 'k--', 
         df.index, df.iloc[:,2], 'b--',
         df.index, df.iloc[:,3], 'k-',
         df.index, df.iloc[:,4], 'g-',
         df.index, df.iloc[:,5], 'b-',)

plt.title("Page views on English Wikipedia (x 10,000,000,000)")
plt.xlabel("Year")
plt.ylabel("Views")
plt.legend(['Old API Total', 'Old API Desktop', 'Old API Mobile', 'New API Total', 'New API Desktop', 'New API Mobile'])
plt.savefig("A1_plot.jpeg")
```
